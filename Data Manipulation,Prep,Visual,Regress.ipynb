```python
import pandas as pd
import os
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import calendar 
from datetime import datetime
%matplotlib inline
```

```python
cwd= os.getcwd()
data  = pd.read_csv(cwd+'/data.csv')
data.head()
```

```python
data.drop(columns=['Unnamed: 0'],inplace = True)
```

```python
data.rename(columns={'Total Profit':'Total_Profit','Total Revenue':'Total_Revenue'},inplace=True)
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data['% margin']=data['Total_Profit']/data['Total_Revenue']*100
```

```python
def Decision(col):
    if col >= 30 :
        return 'good'
    else :
        return 'bad'
```

```python
data['Decision']=data['% margin'].apply(Decision)
```

```python
data
```

```python
sns.scatterplot(x='Total_Revenue',y='Total_Profit', data=data ) #detech abnormal
```

# Regression

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
X=datas[['Total_Revenue','% margin','Total Cost']]
y=datas[['Total_Profit']]
```

```python
X_train, X_test, y_train, y_test =  train_test_split (X,y , train_size=0.7, random_state=1000)
```

```python
model =sm.OLS(y_train,X_train).fit()
```

```python
model.summary()
```

# Prediction

```python
ypred = model.predict(X_test)
```

# Plot comparison

```python
plt.scatter(ypred,y_test)
```
